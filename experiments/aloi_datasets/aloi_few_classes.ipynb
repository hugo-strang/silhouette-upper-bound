```python
from silhouette_upper_bound import upper_bound
import numpy as np
import kmedoids
from scipy.spatial.distance import squareform, pdist
from common import get_df_sampled, eval 
```

```python
def get_distance_matrix_and_labels(df_sampled, n_classes, n_features, seed):
    """
    Help function for loading distance matrix and class labels.
    """

    # set seed
    np.random.seed(seed)

    # sample classes 
    classes = df_sampled["class"].unique()
    chosen = np.random.choice(classes, size=n_classes, replace=False)

    # Filter dataframe to those two classes
    df_sampled_n_classes = df_sampled[df_sampled["class"].isin(chosen)]
    print("Classes: " + str(df_sampled_n_classes["class"].unique()))

    # Separate data into feature vectors and class labels 
    y = df_sampled_n_classes["class"]
    X = df_sampled_n_classes.iloc[:, 0:n_features]
    X = X.to_numpy()

    print(f"Shape of feature vectors: {X.shape}. Data type: {X.dtype}")
    D = squareform(pdist(X, metric="cityblock"))

    return D, y 
```

# 1. ALOI HSB 5x5x5

```python
dataset_name = "aloi-hsb-5x5x5"
N_FEATURES = 125
```

```python
df_sampled = get_df_sampled(dataset_name=dataset_name)
df_sampled.head()
```

### n classes = 2

```python
n_classes = 2

# Load data
D2, y2 = get_distance_matrix_and_labels(df_sampled=df_sampled, n_classes=n_classes, n_features=N_FEATURES, seed=872)
ub = upper_bound(D2)
```

```python
# PAMSIL
eval(D2, y2, (kmedoids.pamsil(diss=D2, medoids=n_classes, random_state=42).labels + 1), ub)
```

```python
# FastMSC
eval(D2, y2, (kmedoids.fastmsc(diss=D2, medoids=n_classes, random_state=42).labels + 1), ub)
```

```python
# DynMSC
eval(D2, y2, (kmedoids.dynmsc(diss=D2, medoids=10, random_state=42).labels + 1), ub)
```
